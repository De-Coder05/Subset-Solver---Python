```python
import random as r
a = set([-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4])
LB=3
UB=6
resultList = set()
iterations = 10000
for i in range(iterations):
    random = r.sample(sorted(a),r.randint(LB,UB))
    random.sort()
    if sum(random) == 0:
        resultList.add(tuple(random))
for ran in resultList:
	print (ran)
print ("\nTotal Sets: ", len(resultList))
```
